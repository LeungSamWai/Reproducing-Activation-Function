```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8,8))
error2 = np.load("error_save_La_2D_w50_relu3.npy")
error2 = error2.reshape(500,100)
error2 = np.average(error2,1)
print(error2[499])
ax.semilogy(np.arange(0,50000,100),error2,label="ReLU$^3$",color = 'tab:orange')
error1 = np.load("error_save_La_2D_w50_poly.npy")
error1 = error1.reshape(500,100)
error1 = np.average(error1,1)
print(error1[499])
plt.semilogy(np.arange(0,50000,100),error1,label="$x\oplus x^2$",color = 'tab:green')


error4 = np.load("error_save_La_2D_w50_poly_relu.npy")

error4 = error4.reshape(500,100)
error4 = np.average(error4,1)
print(error4[499])
plt.semilogy(np.arange(0,50000,100),error4,label="$x\oplus x^2\oplus$ ReLU",color = 'tab:red')

error5 = np.load("error_save_La_2D_w50_poly_relu3.npy")
error5 = error5.reshape(500,100)
error5 = np.average(error5,1)
print(error5[499])
plt.semilogy(np.arange(0,50000,100),error5,label="$x\oplus x^2\oplus$ ReLU$^3$",color = 'tab:purple')

error3 = np.load("error_save_La_2D_w50_poly_sin.npy")
error3 = error3.reshape(500,100)
error3 = np.average(error3,1)
print(error3[499])
plt.semilogy(np.arange(0,50000,100),error3,label="$x\oplus x^2\oplus sin(x)$",color = 'tab:brown')

error3 = np.load("error_save_La_2D_w50_poly_sin_gaussian.npy")
error3 = error3.reshape(1000,100)
error3 = np.average(error3,1)
print(error3[999])
plt.semilogy(np.arange(0,100000,100),error3,label="$x\oplus x^2 \oplus sin(x)\oplus $Gaussian",color = 'tab:pink')
ax.legend(fontsize=20)
ax.set_xticks(np.arange(0,100001,20000))
ax.tick_params(labelsize=30)
plt.ylabel('Relative $L^2$ error',fontsize=30)
plt.xlabel('Iterations numbers',fontsize=30)
plt.subplots_adjust(left=0.21,right=0.95,top=0.9,bottom=0.16)
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.xaxis.set_major_formatter(formatter)

fig.savefig('../report/Figure/Regular.eps',dpi=600,format='eps')
```
